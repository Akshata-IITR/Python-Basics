```python
# Print all even numbers between given two numbers

a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
Even=[]
Odd=[]
for i in range(a,b+1):
    if(i%2==0):
        Even.append(i)
    else:
        Odd.append(i)
print("Even numbers: ", Even)
print("Odd numbers: ",Odd)
```
